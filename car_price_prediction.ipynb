```python
df = pd.read_csv('Cleaned_Car_data.csv').drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
df.sample(5)
```

```python
X = df.drop(columns='Price')
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=433)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
from warnings import catch_warnings
column_transform = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                           remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_transform,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
